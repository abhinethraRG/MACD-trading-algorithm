### MACD vs Buy and Hold Strategies - Analysis

```python
import pandas as pd
```

```python
updated_data = pd.read_excel("C:\\SMU\\winter term\\Acct & Financial Analytics\\assignments\\A2\\Y2024_MACD_Data - Sol7.xlsx",'Solution')
updated_data.tail()
```

```python
# Filter the dataset to start from the first trading day on or after 01-March-2023
updated_data_filtered = updated_data[updated_data['Date'] >= pd.to_datetime('2023-03-01')].reset_index(drop=True)
```

```python
updated_data_filtered
```

```python
# Initialize variables for trading strategy
owned_stock = 0
buy_price = 0
total_profit = 0
buy_and_hold_profit = 0
```

```python
# Initialize variables
owned_stock = False
buy_price = 0.0
profits = []

# Iterate through the dataset to simulate the trading strategy
for index, row in updated_data_filtered.iterrows():
    action = row.get('Actions', '')
    price = row['Price']

    if action == 'BUY' and not owned_stock:
        buy_price = price
        owned_stock = True
    elif action == 'SELL' and owned_stock:
        sell_price = price
        profits.append(sell_price - buy_price)
        owned_stock = False

# Calculate total profit from the MACD trading strategy
total_trading_profit = sum(profits)

# Calculate profit through the Buy-and-Hold strategy
buy_and_hold_profit = updated_data_filtered.iloc[-1]['Price'] - updated_data_filtered[updated_data_filtered['Date'] == '2023-03-01']['Price'].values[0]

total_trading_profit, buy_and_hold_profit
```
